```python
#treino

!pip install ucimlrepo

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
import pickle as p1

n_neighbors = 15


# import some data to play with
from ucimlrepo import fetch_ucirepo

# fetch dataset
optical_digits = fetch_ucirepo(id=80)

# data (as pandas dataframes)
X = optical_digits.data.features
y = optical_digits.data.targets


n=len(X)
print(n)

X_treino=X[0:int(n*0.7)]
y_treino=y[0:int(n*0.7)]
X_teste=X[int(n*0.7):]
y_teste=y[int(n*0.7):]


with open('X_teste', 'wb') as f:
    p1.dump(X_teste, f)
print("X_teste saved to file.")

with open('y_teste', 'wb') as f:
    p1.dump(y_teste, f)
print("y_teste saved to file.")


KNN=neighbors.KNeighborsClassifier(n_neighbors, weights="uniform")

ModeloKNN=KNN.fit(X_treino, y_treino)

with open('ModeloKNN', 'wb') as preditor_Pickle:
    p1.dump(ModeloKNN, preditor_Pickle)
print("ModeloKNN saved to file.")
```

```python
#Teste
import pickle as p1
import numpy as np
# import pandas as pd # Not strictly needed if X_test/y_test are loaded as DataFrames
# from sklearn import linear_model # Not needed for KNN

# Carregar os dados X_teste
with open('X_teste', 'rb') as f:
    digits_data_X_test = p1.load(f)

# Carregar os dados y_teste
with open('y_teste', 'rb') as f:
    digits_data_Y_test = p1.load(f)

print(type(digits_data_X_test))

# Carregar o modelo treinado
with open('ModeloKNN', 'rb') as f:
    digits_model_loaded = p1.load(f)

# KNeighborsClassifier does not have a .coef_ attribute, typically found in linear models.
# print("Coefficients (digits): \n", digits_model_loaded.coef_)

digits_y_pred = digits_model_loaded.predict(digits_data_X_test)

# Arredondar previsões para o inteiro mais próximo para comparar com as classes reais dos dígitos
# KNN predict already returns integer classes, and it's a 1D array. Flatten ensures 1D.
digits_y_pred_flat = digits_y_pred.flatten()

right_digits = 0
wrong_digits = 0
total_digits = len(digits_data_Y_test)

# Assumindo que digits_data_Y_test é um DataFrame com uma única coluna 'class'
for i in range(total_digits):
    if digits_y_pred_flat[i] == digits_data_Y_test.iloc[i, 0]:
        right_digits += 1
    else:
        wrong_digits += 1

print("Accuracy (digits)= ", right_digits / total_digits)
print("Error Rate (digits)= ", wrong_digits / total_digits)

print("digits ", digits_model_loaded.score(digits_data_X_test,digits_data_Y_test))
```
