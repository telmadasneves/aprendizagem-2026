```python
# =====================================================
# ATIVIDADE 1 - Limpeza dos dados
# =====================================================

import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/Fer412/Dataset_2026/refs/heads/main/heart1.csv"

# nomes das colunas do dataset Heart
colunas = [
    'age','sex','cp','trestbps','chol','fbs',
    'restecg','thalach','exang','oldpeak',
    'slope','ca','thal','target'
]

df = pd.read_csv(url, sep=';', header=None, names=colunas)

print("Primeiras linhas:")
print(df.head())

# Verificar missing values
print("\nMissing values:")
print(df.isnull().sum())

# Caso existam missing values
for coluna in df.columns:
    if df[coluna].isnull().sum() > 0:
        df[coluna].fillna(df[coluna].median(), inplace=True)

# Verificar tipos
print("\nTipos:")
print(df.dtypes)

# Verificar valores não numéricos
for coluna in df.columns:
    if df[coluna].dtype == 'object':
        from sklearn.preprocessing import LabelEncoder
        le = LabelEncoder()
        df[coluna] = le.fit_transform(df[coluna])

# Gravar dataset limpo
df.to_csv("heart_clean_1.csv", index=False)

print("\nheart_clean_1.csv criado.")
```

```python
# =====================================================
# ATIVIDADE 2 - Normalização
# =====================================================

from sklearn.preprocessing import StandardScaler

data = pd.read_csv("heart_clean_1.csv")

X = data.iloc[:, :-1]
y = data.iloc[:, -1]

scaler = StandardScaler()

X_scale = scaler.fit_transform(X)

heart_scale = pd.DataFrame(X_scale, columns=X.columns)
heart_scale['target'] = y

heart_scale.to_csv("heart_scale_2.csv", index=False)

print("heart_scale_2.csv criado.")
```

```python
# =====================================================
# ATIVIDADE 3 - Train/Test Split
# =====================================================

from sklearn.model_selection import train_test_split

data = pd.read_csv("heart_scale_2.csv")

treino, teste = train_test_split(
    data,
    test_size=0.20,
    random_state=21
)

treino.to_csv("heart_treino_3.csv", index=False)
teste.to_csv("heart_teste_3.csv", index=False)

print("heart_treino_3.csv criado.")
print("heart_teste_3.csv criado.")
```

```python
# =====================================================
# ATIVIDADE 4 - Treino da RNA
# =====================================================

from sklearn.neural_network import MLPClassifier
import pickle

train = pd.read_csv("heart_treino_3.csv")

X_train = train.iloc[:, :-1].values
y_train = train.iloc[:, -1].values

classifier = MLPClassifier(
    hidden_layer_sizes=(150,100,50),
    max_iter=100000,
    activation='relu',
    solver='adam',
    random_state=1
)

classifier.fit(X_train, y_train)

pickle.dump(classifier, open("heart_model.sav", "wb"))

print("Modelo treinado e guardado.")
```

```python
# =====================================================
# ATIVIDADE 5 - Avaliação
# =====================================================

from sklearn.metrics import confusion_matrix, accuracy_score
import pickle

modelo = pickle.load(open("heart_model.sav", "rb"))

teste = pd.read_csv("heart_teste_3.csv")

X_test = teste.iloc[:, :-1].values
y_test = teste.iloc[:, -1].values

y_pred = modelo.predict(X_test)

score = modelo.score(X_test, y_test)

cm = confusion_matrix(y_pred, y_test)

print("Score:", score)
print("\nMatriz de Confusão:")
print(cm)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
# =====================================================
# ATIVIDADE 6 - Predição
# =====================================================

import pickle
import pandas as pd

modelo = pickle.load(open("heart_model.sav", "rb"))

print("Introduza os 13 valores separados por vírgulas:")

entrada = input()

valores = [float(x) for x in entrada.split(",")]

df = pd.DataFrame([valores])

resultado = modelo.predict(df)

print("Predição:", resultado[0])

if resultado[0] == 1:
    print("Doença cardíaca detectada")
else:
    print("Sem doença cardíaca")
```
